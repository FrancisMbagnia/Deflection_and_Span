```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from ipywidgets import interact
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
def cauchy(Z,t):
    z0=Z[0]
    z1=Z[1]
    z2=Z[2]
    z3=Z[3]
    z0prime=z1
    z1prime=z2
    z2prime=z3
    z3prime=-t*z3+z1**2-t**2*np.sin(z0)
    Zprime=np.array([z0prime,z1prime,z2prime,z3prime])
    return Zprime

Z0=np.array([0,1,2,-1])
tt=np.arange(2,7,0.00002)
print(tt.shape)
Zode=odeint(cauchy,Z0,tt)
print(Zode.shape)
istart=int((2.51-2)/0.00002)
ifinish=int((5-2)/0.00002)
print(istart)
print(ifinish)
print(tt[istart])
print(tt[ifinish])
plt.plot(tt[istart:ifinish],Zode[istart:ifinish,2])
```

```python
def cauchy2(Z,t):
    w0=Z[0]
    w1=Z[1]
    z0=Z[2]
    z1=Z[3]
    w0prime=w1
    w1prime=-3*w1+z1-4*z0
    z0prime=z1
    z1prime=-5*z1+2*w1-w0
    Zprime=np.array([w0prime,w1prime,z0prime,z1prime])
    return Zprime

Z02=np.array([0,-1,0,1])
t2=np.arange(1,8,0.0005)
Zode2=odeint(cauchy2,Z02,t2)
istart2=int((2.3-1)/0.0005)
ifinish2=int((6.41-1)/0.0005)
print(istart2)
print(ifinish2)
print(t2[istart2])
print(t2[ifinish2])
plt.plot(t2[istart2:ifinish2],Zode2[istart2:ifinish2,2])
plt.figure()
plt.plot(Zode2[istart2:ifinish2,0],Zode2[istart2:ifinish2,1])
```
